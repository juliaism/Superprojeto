```python
import pandas as pd
import numpy as np

df = pd.read_csv('data/Student_Dropout.csv', sep=';', lineterminator='\r')

df.head()
```

```python
# Tirando a linha vazia do arquivo que tava gerando valor nulo em todas as colunas
df = df.dropna().copy()
```

```python
# Removendo os alunos que ainda estão matriculados e deixando só quem se formou ou desistiu
df = df[df['Target'].isin(['Graduate', 'Dropout'])].copy()
```

```python
# Criando a coluna em português
df['situacao'] = df['Target'].map({'Graduate': 'Sucesso', 'Dropout': 'Abandono'})
```

```python
# Achatando a idade de quem entrou muito velho para o limite de 34 anos usando o IQR
q1, q3 = df['Age at enrollment'].quantile([0.25, 0.75])
limite_superior = q3 + (1.5 * (q3 - q1))
df['idade_ingresso'] = np.where(df['Age at enrollment'] > limite_superior, limite_superior, df['Age at enrollment'])
```

```python
# Traduzindo 
df['bolsista'] = df['Scholarship holder'].map({1: 'Sim', 0: 'Não'})
df['turno'] = df['Daytime/evening attendance\t'].map({1: 'Diurno', 0: 'Noturno'})
df['sexo'] = df['Gender'].map({1: 'Masculino', 0: 'Feminino'})
```

```python
# Deletando colunas que foram traduzidas
df = df.drop(['Daytime/evening attendance\t', 'Scholarship holder', 'Gender', 'Target'], axis=1)
```

```python
df.to_csv('data/estudantes_limpo.csv', index=False)
```
